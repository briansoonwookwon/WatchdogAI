```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train_losses.csv')
eval = pd.read_csv('eval_losses.csv')
```

```python
plt.plot(train["train/global_step"],train["output-models - train/loss"])
plt.plot(eval["train/global_step"], eval["output-models - eval/loss"])
plt.title('ViT Model Loss (AI Dectection)')
plt.ylabel('Loss')
plt.xlabel('Steps')
plt.legend(['Training', 'Validation'], loc='upper right')
plt.show()
```
